```python
import pandas as pd
#Now we will load our data 
df = pd.read_csv("titanic.csv")

#Here we are finding out the number of rows and column present in our data
print("shape:", df.shape)

#Here we print the 1st 5 rows of our dataset
df.head()
```

```python
#checking for presence of any null values
df.isnull().sum()
```

```python
# Lets categorise the age according to pclass and then find their medican and replace accordingly
df['Age'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
#in the result we can see we have rectified all the missing age values
```

```python
df['Fare'].fillna(df['Fare'].median(), inplace=True)
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
df['Title'] = df['Title'].replace(['Mlle', 'Ms'], 'Miss').replace(['Mme'], 'Mrs')
```
